# Scraping NewEgg for information on monitors
Building a Database

```python
from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
```

```python
url = 'https://www.newegg.com/4k-2k-Monitors/PromotionStore/ID-2032604?cm_sp=CAT_Monitors-_-VisNav_2-_-4K-42-Monitors_1'
```

```python
content = urllib.request.urlopen(url)
soup = BeautifulSoup(content, 'html.parser')
```

```python
soup
```

## What we want from the page
-Make pandas dataframe, save dataframe to csv
Brand 
Price 
Size
Free shipping (bool)
Student price (bool)
Resolution
Rating
Speakers built in (bool)

```python
# we need to get the div for each item
# <div class="item-container  ">
```

```python
# the products are in div "item-container"
products = soup.find_all('div', class_="item-container")
```

```python
len(products)
# an estimate of the number of products listed on the page
```

```python
# lets parse the first div (i.e. the first product on the page) 
# to get an idea of how to parse the rest of the divs
first_div = products[0]
first_div
```

```python
# we saw a lot of information in the img tag so let's use that tag to extract the brand
image = first_div.find_all('img')
```

```python
image_s = str(image[0])
image_s
```

```python
# brand is under title
# get the brand name
title_div1 = image_s.split('title')
title = title_div1[1][2:]
title.split(' ', 1)
```

```python
# it looks like we can get the size of the screen from the img element as well
size_test = title[1]
size = size_test.split('"')[0].split(' ')[1]
size
```

```python
# we need lists to APPEND the data we find
# we create empty lists that we will append to in the loop below
all_brands = []
all_sizes = []
all_prices = []

# we can loop over each product in the page (each product is in a div "item-container" as we found above)
# for each product we will find the brand, size, and price and append each to their respective lists initialized above
for product in products:
    # find the brand
    image = product.find_all('img')
    image_s = str(image[0])
    title_div1 = image_s.split('title')
    title = title_div1[1][2:]
    title = title.split(' ', 1)
    # print(title[0])
    # append brand to list all_brands
    all_brands.append(title[0])
    # find the size
    size_div = title[1]
    size = size_div.split('"')[0].split(' ')[1]
    #print(size)
    all_sizes.append(size)
    # get price
    price_tag = product.find_all('strong')[-1]
    price_tag.text
    # print(price_tag.text)
    all_prices.append(price_tag.text)
```

```python
# We printed each list below to check how well our program mined the size, brand, and price automatically
# It looks like we did okay, but there are a few bugs. 
# Can you find where we can fix this?
```

```python
# create a dataframe from our 3 lists
monitordata = pd.DataFrame(
    {'price': all_prices,
     'brand': all_brands,
     'size': all_sizes
    })
```

```python
# save the dataframe as a csv locally
monitordata.to_csv('monitordata.csv')
```
